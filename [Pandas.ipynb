```python
# Pandas Library:
# Useful for Data Processing & Analysis

# Pandas Data Frame:
# Two-dimensional tabular data structure with labeled axes (rows and columns).

# Importing libraries
import pandas as pd
import numpy as np
```

```python
# üè° Loading California Housing Dataset (modern replacement for Boston Housing)
from sklearn.datasets import fetch_california_housing

# Load dataset as a pandas DataFrame
housing_dataset = fetch_california_housing(as_frame=True)

# Convert to DataFrame
boston_df = housing_dataset.frame

boston_df.head()
```

```python
# Check shape and type
print("Shape of dataset:", boston_df.shape)
print("Type of dataset:", type(boston_df))
```

```python
# Statistical measures for California Housing dataset
print("Count of values per column:")
print(boston_df.count())

print("\nMean values per column:")
print(boston_df.mean())

print("\nStandard deviation per column:")
print(boston_df.std())

print("\nMinimum values per column:")
print(boston_df.min())

print("\nMaximum values per column:")
print(boston_df.max())

print("\nFull summary statistics:")
boston_df.describe()
```

```python

boston_df['MedHouseValue'] = housing_dataset.target

boston_df.head()
```

```python

boston_df.drop(index=0, axis=0, inplace=False)  # inplace=False keeps original DataFrame

boston_df.drop(columns='HouseAge', axis=1, inplace=False)
```

```python
# Locate a specific row
print("Row at index 2:")
print(boston_df.iloc[2])

# Locate particular columns
print("\nFirst column:")
print(boston_df.iloc[:, 0])

print("\nSecond column:")
print(boston_df.iloc[:, 1])

print("\nThird column:")
print(boston_df.iloc[:, 2])

print("\nLast column:")
print(boston_df.iloc[:, -1])
```

```python
# Correlation: positive and negative correlations
corr_matrix = boston_df.corr()
print(corr_matrix)

# Optional: display correlation heatmap (requires seaborn)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap for California Housing Dataset")
plt.show()
```
